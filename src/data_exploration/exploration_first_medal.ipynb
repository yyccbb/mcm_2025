```python
import pandas as pd
import numpy as np
```

```python
athletes = pd.read_csv('../data/summerOly_athletes.csv')
athletes.shape[0]
```

```python
df_before_2024 = athletes[athletes['Year'] < 2024]
```

```python
df_before_2024
```

```python
nocs_no_medals = df_before_2024.groupby('NOC').filter(lambda group: (group['Medal'] == "No medal").all())
```

```python
nocs_no_medals
```

```python
len(nocs_no_medals['Team'].unique())
```

```python
nocs_no_medals_after_2024 = athletes.groupby('NOC').filter(lambda group: (group['Medal'] == "No medal").all())
```

```python
nocs_no_medals_after_2024
```

```python
nocs_no_medals_after_2024['Team'].unique()
```

```python
first_medals_2024 = np.setdiff1d(nocs_no_medals['Team'].unique(), nocs_no_medals_after_2024['Team'].unique())
```

```python
no_medals_emerging_2024 = np.setdiff1d(nocs_no_medals_after_2024['Team'].unique(), nocs_no_medals['Team'].unique())
```

```python
len(nocs_no_medals_after_2024['Team'].unique())
```

```python
filtered_ds = athletes[athletes['Team'].isin(no_medals_emerging_2024)]
filtered_ds
```

```python
emerging_2024 = np.setdiff1d(athletes['Team'].unique(), df_before_2024['Team'].unique())
emerging_2024
```

```python
# Past participation records for countries that got their first medals in 2024
NOCs_first_medal_2024 = athletes[athletes['Team'].isin(first_medals_2024)]
NOCs_first_medal_2024
```

```python
for noc in first_medals_2024:
    participating_years = athletes[athletes['Team'] == noc]['Year'].unique()
    print(f"{noc} first participated in {participating_years.min()} and got their first medal in 2024")
```
